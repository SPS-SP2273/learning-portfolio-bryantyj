```python
import numpy as np

f= lambda x: np.cos(x) - x                             #f can be defined for any function

def get_root(x_left, x_right):                         # function which takes in upper and lower bound of x
                                                       # x_left < x_right
    x_midpoint = (x_left + x_right) / 2                # x-value of midpoint of upper and lower bound of x
    x_left_answer = f(x_left)                          # y-value of lower bound
    x_right_answer = f(x_right)                        # y-value of upper bound
    x_midpoint_answer = f(x_midpoint)                  # y-value of mid-point
    if abs(x_left - x_right) < 1e-10:                  # difference between left and right bound is similar to 10dp
        return round(x_midpoint, 10)                   # when 10dp similarity return the answer
    else:                                              # IF NOT 10DP SIMILARITY,
        if (x_left_answer * x_midpoint_answer) > 0:    # if left bound same sign as midpoint,
            return get_root(x_midpoint, x_right)       # reduce bound to (midpoint, upper bound) AND rerun get_root function
        else:                                          # if not,(left bound different sign as midpoint/ right point same sign as midpoint)
            return get_root(x_left, x_midpoint)        # reduce bound to (lower bound, midpoint) AND rerun get_root function
                                                       
get_root(0.5, 0.9)
```

*Intermediate Value Theorem*:  

This code can be generalised form where we are not just finding root at 0  

LIMITATION:  
1. when f(upper bound) OR f(lower bound) OR f(midpoint) == 0,  
   then there will be an error since it is the solution already solved when y-value == 0  
   
2. root must be between upper and lower bound else will return upper bound due to the else syntax
   
2. MUST have bracketing value
   eg, y=x^2 will not work
  
3. Slow convergence rate (linear rate of conversion)
  
4. Typically bisection is used to get an initial estimate for such faster methods 
   such as Newton-Raphson that requires an initial estimate. 
   There is also the inability to detect multiple roots.  
   
5. Cannot detect multiple roots/ complex roots

6. Function must be continuous, else will have undefined values

7. Choosing one bound to be close to root will not make it faster, still might take many iterations
