```python
import numpy as np

    
x=[]
y=[]

for i in range (51):
    x.append(i) 
x=np.linspace(0,51,1000)
for j in x: 
    output=np.exp(-(j/10))*np.sin(j)
    y.append(output)
    
plt.plot(x, y, color='cornflowerblue')
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15, rotation=0)
plt.ylim(-1,1)
plt.title(r'$y=e^{-x/10}sin(x)$', fontsize=20)
plt.grid(alpha=.25)
plt.tick_params(axis='x', color= 'darkblue')    #colors=number only, color=axis dashes+number
plt.tick_params(axis='y', color= 'darkblue')
```
